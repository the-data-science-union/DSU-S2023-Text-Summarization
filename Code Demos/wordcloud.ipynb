```python
from datasets import load_dataset

# choose what split you want from the dataset using the split argument
dataset = load_dataset("billsum", split = "test")
```

```python
# load dataset as pandas dataframe
import pandas as pd
dataset.set_format("pandas")
df = dataset[0:]
df
```

```python
# import necessary libraries 
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
```

```python
# generate sameple wordcloud visualization with just first row
sample_text = df["text"][1]
wordcloud = WordCloud(max_words = 70, max_font_size = 60, background_color = "white").generate(sample_text)

plt.figure(figsize = (8,8))
# interpolation argument is simply a method to make the image appear smoother
plt.imshow(wordcloud, interpolation='bilinear') 
plt.axis("off")
plt.show()

# to save image:
# wordcloud.to_file("img/first_review.png")
```

```python
# make wordcloud viz with first 1000 rows
stopwords = ["\n", "SHORT TITLE", "SECTION 1"]

text = "".join(row for row in df["text"][1:1000])
wordcloud = WordCloud(stopwords = stopwords, max_words = 70, max_font_size = 60, background_color = "white").generate(text)

plt.figure(figsize = (8,8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
